```python
X_train=X_train
X_test=X_test
```

```python
def model(w,b,x):
    if(np.dot(w,x)>=b):
        return 1
    else:
        return 0

def predict(w,b,X):
    Y=[]
    for x in X:
        result=model(w,b,x)
        Y.append(result)
    return np.array(Y)

def fit(X,Y,epochs=1,lr=1):
    w=np.ones(X.shape[1])
    b=0
    
    accuracy={}
    max_accuracy=0
    
    wt_matrix=[]
    
    for i in range(epochs):
        for x,y in zip(X,Y):
            y_pred =model(w,b,x)
            if y==1 and y_pred==0:
                w=w+lr*x
                b=b-lr*1
            elif y==0 and y_pred==1:
                w=w-lr*x
                b=b+lr*1
            
        wt_matrix.append(w)
        
        accuracy[i]=accuracy_score(predict(w,b,X),Y)
        if(accuracy[i]>max_accuracy):
            max_accuracy=accuracy[i]
            chkptw=w
            chkptb=b
        
        w=chkptw
        b=chkptb
        
    print(max_accuracy)
    plt.plot(accuracy.values())
    plt.ylim([0,1])
    plt.show()
        
    return np.array(wt_matrix),w,b
    
```

```python
X_train = X_train.astype(float)
import numpy as np
X_train = np.array(X_train, dtype=float)
```
